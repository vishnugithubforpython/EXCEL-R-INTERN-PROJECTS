```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
km
```

```python
x
```

```python
km.fit(x)
```

```python
cluster_labels=km.labels_
cluster_labels
```

```python
# Visualize the clustering results (for 2D data)
# Assuming you choose two features for visualization (you can choose any two)
feature1 = 'Alcohol'
feature2 = 'Color'

plt.figure(figsize=(10, 6))
plt.scatter(x[feature1], x[feature2], c=cluster_labels, cmap='viridis')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title('K-means Clustering')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Evaluate clustering performance using silhouette score
silhouette = silhouette_score(x, cluster_labels)
print("Silhouette Score:", silhouette)

# Evaluate clustering performance using Davies-Bouldin index
davies_bouldin = davies_bouldin_score(x, cluster_labels)
print("Davies-Bouldin Index:", davies_bouldin)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 0], c=cluster_pca, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 1')
plt.title('K-means Clustering on PCA-transformed Data')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Visualize side by side (Original vs. PCA-transformed)
plt.figure(figsize=(15, 6))

# Original Data
plt.subplot(1, 2, 1)
plt.scatter(x[feature1], x[feature2], c=cluster_labels, cmap='viridis')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title('K-means Clustering (Original Data)')
plt.colorbar(label='Cluster')

# PCA-transformed Data
plt.subplot(1, 2, 2)
plt.scatter(x_pca[:, 0], x_pca[:, 0], c=cluster_pca, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering (PCA-transformed Data)')
plt.colorbar(label='Cluster')

plt.tight_layout()
plt.show()
```

```python
Similarities: Both approaches aim to group similar wine samples together based on their characteristics. 
In both cases, we obtain clusters representing different types of wine.
Differences: The clusters obtained from the PCA-transformed data are
in a reduced-dimensional space defined by the principal components. 
This means that the clusters are formed based on combinations of the original features,
potentially resulting in different cluster shapes and orientations compared to clustering on the original dataset.
```

```python
PCA:
Pros: Reduces computational complexity by working with fewer dimensions, especially beneficial for high-dimensional datasets. 
Can reveal underlying patterns in the data and highlight important features
```

```python
Clustering on Original Data:
Pros: Utilizes the original feature space, preserving all available information. Clusters may be more interpretable in terms of the original features.
```
