```python
numerical_columns=["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]
result_dict={}
for column in numerical_columns:
    mean=df[column].mean()
    median=df[column].median()
    mode=df[column].mode()
    std_dev=df[column].std()
    result_dict[column] = {'Mean': mean, 'Median': median, 'Mode': mode, 'Standard Deviation': std_dev}
result_df=pd.DataFrame(result_dict)
print(result_df)
```

```python
numerical_columns = ["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]
Q1 = df[numerical_columns].quantile(0.25)
Q3 = df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ((df[numerical_columns] < lower_bound) | (df[numerical_columns] > upper_bound))

print("\nOutliers:")
print(outliers.sum())


```

```python

categorical_columns = ["Date", "Day", "SKU", "City", "Volume", "BU", "Brand", "Model"]
for column in categorical_columns:
    category_counts = df[column].value_counts()
    plt.figure(figsize=(10, 6))
    category_counts.plot(kind='bar')
    plt.title(f'Frequency of Categories in {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)  
    plt.show()

    print(f"Top categories in {column}:")
    print(category_counts.head())
    print("\n")
```

```python
 numerical_columns = ["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for column in numerical_columns:
    df[column].hist(alpha=0.5, label=column)
plt.title('Before Standardization')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

for column in numerical_columns:
    mean = df[column].mean()
    std_dev = df[column].std()
    df[column] = (df[column] - mean) / std_dev
plt.subplot(1, 2, 2)
for column in numerical_columns:
    df[column].hist(alpha=0.5, label=column)
plt.title('After Standardization')
plt.xlabel('Values (Standardized)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
