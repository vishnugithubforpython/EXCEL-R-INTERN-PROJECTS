```python

from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.svm import SVR
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
X = x_train
y = y_train


# Define the list of kernels to compare
kernels = ['linear', 'poly', 'rbf']

# Define a dictionary to store performance metrics for each kernel
performance_metrics = {
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1_score': []
}

# Define scoring functions for precision, recall, and f1-score
scoring = {
    'accuracy': 'accuracy',
    'precision': make_scorer(precision_score, average='weighted'),
    'recall': make_scorer(recall_score, average='weighted'),
    'f1_score': make_scorer(f1_score, average='weighted')
}

# Perform cross-validation for each kernel
for kernel in kernels:
    # Create SVM model with the specified kernel
    model = SVR(kernel=kernel)
    
    # Perform cross-validation with 5 folds for all metrics
    scores = cross_validate(model, X, y, cv=5, scoring=scoring)
    
    # Store mean performance metrics for this kernel
    performance_metrics['accuracy'].append(np.mean(scores['test_accuracy']))
    performance_metrics['precision'].append(np.mean(scores['test_precision']))
    performance_metrics['recall'].append(np.mean(scores['test_recall']))
    performance_metrics['f1_score'].append(np.mean(scores['test_f1_score']))

# Print performance metrics for each kernel
print("Performance metrics for SVM with different kernels:")
for kernel in kernels:
    print(f"Kernel: {kernel}")
    print(f"  Accuracy: {performance_metrics['accuracy'][kernels.index(kernel)]:.2f}")
    print(f"  Precision: {performance_metrics['precision'][kernels.index(kernel)]:.2f}")
    print(f"  Recall: {performance_metrics['recall'][kernels.index(kernel)]:.2f}")
    print(f"  F1-score: {performance_metrics['f1_score'][kernels.index(kernel)]:.2f}")
    print()
```
