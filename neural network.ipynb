```python
# Encode the target labels
label_encoder = LabelEncoder()
data['letter'] = label_encoder.fit_transform(data['letter'])
```

```python
X = data.drop('letter', axis=1)
y = data['letter']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the ANN model
def create_model(input_dim, output_dim, layers=1, units=64, activation='relu', optimizer='adam'):
    model = Sequential()
    model.add(Dense(units, input_dim=input_dim, activation=activation))
    for _ in range(layers - 1):
        model.add(Dense(units, activation=activation))
    model.add(Dense(output_dim, activation='softmax'))
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

input_dim = X_train.shape[1]
output_dim = len(label_encoder.classes_)

# Create and compile the model
model = create_model(input_dim, output_dim)
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy:.2f}')

```

```python
def tune_hyperparameters(X_train, y_train, X_test, y_test):
    results = []
    input_dim = X_train.shape[1]
    output_dim = len(label_encoder.classes_)

    # Define hyperparameter grid
    layers_options = [1, 2, 3]
    units_options = [32, 64, 128]
    activation_options = ['relu', 'tanh']
    optimizer_options = ['adam', 'sgd']
    epochs_options = [10, 20]
    batch_size_options = [32, 64]

    # Iterate over all combinations of hyperparameters
    for layers in layers_options:
        for units in units_options:
            for activation in activation_options:
                for optimizer in optimizer_options:
                    for epochs in epochs_options:
                        for batch_size in batch_size_options:
                            model = create_model(input_dim, output_dim, layers=layers, units=units, activation=activation, optimizer=optimizer)
                            history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=0)
                            test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
                            results.append((test_accuracy, layers, units, activation, optimizer, epochs, batch_size))
                            print(f'Accuracy: {test_accuracy:.2f}, Layers: {layers}, Units: {units}, Activation: {activation}, Optimizer: {optimizer}, Epochs: {epochs}, Batch Size: {batch_size}')
    
    # Find the best hyperparameters
    if results:
        best_params = max(results, key=lambda item: item[0])
        print(f'Best Accuracy: {best_params[0]:.2f}')
        print(f'Best Parameters: Layers: {best_params[1]}, Units: {best_params[2]}, Activation: {best_params[3]}, Optimizer: {best_params[4]}, Epochs: {best_params[5]}, Batch Size: {best_params[6]}')
        return best_params
    else:
        print('No hyperparameter combinations were evaluated.')
        return None

# Run hyperparameter tuning and get the best parameters
best_params = tune_hyperparameters(X_train, y_train, X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Use the best hyperparameters to create the final model
best_model = create_model(input_dim, output_dim, layers=best_params[1], units=best_params[2], activation=best_params[3], optimizer=best_params[4])
best_model.fit(X_train, y_train, epochs=best_params[5], batch_size=best_params[6], validation_data=(X_test, y_test), verbose=0)

# Predictions with the best model
y_pred = best_model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)

# Evaluation metrics
print(classification_report(y_test, y_pred_classes, target_names=label_encoder.classes_))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)
print(conf_matrix)
```
