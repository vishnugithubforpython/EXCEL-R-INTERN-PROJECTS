```python
def precision_recall_f1(ground_truth, recommendations):
    if not recommendations:
        return 0.0, 0.0, 0.0
    
    true_positives = len(set(ground_truth) & set(recommendations))
    false_positives = len(set(recommendations) - set(ground_truth))
    false_negatives = len(set(ground_truth) - set(recommendations))
    
    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0.0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if precision + recall > 0 else 0.0
    
    return precision, recall, f1
```
