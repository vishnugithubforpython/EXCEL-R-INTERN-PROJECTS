```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("C://Users//mvish//Desktop//Cardiotocographic.csv")
df
```

```python
missing_values = df.isnull().sum()

print("Missing Values:")
print(missing_values)
```

```python
filled_data=df.fillna(0)
filled_data
```

```python
new_values=filled_data.isnull().sum()
new_values
```

```python
datatypes=filled_data.dtypes
datatypes
```

```python
filled_data.describe()
```

```python
#Histogram
filled_data.hist(figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
# Boxplots
filled_data.plot(kind='box', figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
categorical_columns = ["LB","AC","FM","UC","DL","DS","DP","ASTV","MSTV","ALTV","MLTV","Width","Tendency","NSP"]
for column in categorical_columns:
    category_counts = filled_data[column].value_counts()
    plt.figure(figsize=(10, 8))
    category_counts.plot(kind='bar')
    plt.title(f'Frequency of Categories in {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45) 
    plt.tight_layout()
    plt.show()

    print(f"Top categories in {column}:")
    print(category_counts.head())
    print("\n")
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(filled_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Generate scatter plots for select pairs of variables
sns.pairplot(filled_data, vars=['LB', 'AC', 'FM', 'UC', 'DL'], kind='scatter')
plt.suptitle('Pair Plot for Selected Variables', y=1.02)
plt.show()

# Employ advanced visualization techniques like violin plots
plt.figure(figsize=(12, 8))
sns.violinplot(x='ALTV', y='LB', filled_data=filled_data)
plt.title('Violin Plot of ALTV vs LB')
plt.show()
```
