```python
er_education = OneHotEncoder(sparse=False)
er_race = OneHotEncoder(sparse=False)
er_sex = OneHotEncoder(sparse=False)
er_native_country = OneHotEncoder(sparse=False)

# Fit and transform each categorical column
encoded_education = er_education.fit_transform(new[['education']])
encoded_race = er_race.fit_transform(new[['race']])
encoded_sex = er_sex.fit_transform(new[['sex']])
encoded_native_country = er_native_country.fit_transform(new[['native_country']])

# Get the feature names for the encoded columns
encoded_education_df = pd.DataFrame(encoded_education, columns=er_education.get_feature_names_out(['education']))
encoded_race_df = pd.DataFrame(encoded_race, columns=er_race.get_feature_names_out(['race']))
encoded_sex_df = pd.DataFrame(encoded_sex, columns=er_sex.get_feature_names_out(['sex']))
encoded_native_country_df = pd.DataFrame(encoded_native_country, columns=er_native_country.get_feature_names_out(['native_country']))

# Drop the original categorical columns and concatenate the new encoded columns
new = new.drop(columns=['education', 'race', 'sex', 'native_country'])
new = pd.concat([new, encoded_education_df, encoded_race_df, encoded_sex_df, encoded_native_country_df], axis=1)

print(new)
```

```python

# Sample DataFrame
data = {
    'age': [25, 32, 40, 28, 35],
    'income': [50000, 60000, 70000, 55000, 65000]
}

df = pd.DataFrame(data)

# Create Age Group feature
bins = [0, 25, 35, 45, 100]  # Define age group bins
labels = ['Young', 'Young Adult', 'Adult', 'Senior']  # Define age group labels
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Create Income-to-Age Ratio feature
df['income_age_ratio'] = df['income'] / df['age']

print(df)
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame with skewed numerical feature (income)
data = {
    'age': [25, 32, 40, 28, 35],
    'income': [50000, 60000, 70000, 55000, 65000]
}

df = pd.DataFrame(data)

# Apply log transformation to skewed numerical feature (income)
df['log_income'] = np.log1p(df['income'])  # Use log1p to avoid log(0)

print(df)
```
